```python
import pandas as pd
```

```python
# Загрузка данных 
df  = pd.read_csv(r"clicks_dataset_msk_20230101_20230725_spec.csv")

df.columns = ['zapros','object','result']

df
```

```python
# Отбираем объекты с отметкой 1
obj_1 = df[df.result == 1]
d_1 =  obj_1.groupby('object', as_index = False).agg({'zapros':'nunique'})

d_1.columns = ['object','result_1']
d_1
```

```python
# Отбираем запросы с меткой 0 
obj_2 = df[df.result == 0]

d_2 =  obj_2.groupby('object', as_index = False).agg({'zapros':'nunique'})

d_2.columns = ['object','result_0']
d_2
```

```python
# Объединяем датафреймы с 1 и 0 
res = d_1.merge(d_2, on = 'object', how = 'left')

res = res.fillna(0)
```

```python
# Разница в кол-ве запросов
res['disbalance'] = abs(res.result_1 - res.result_0)
res
```

```python
# Дисбаланс 1 и 0 
res.disbalance.value_counts().reset_index()
```
